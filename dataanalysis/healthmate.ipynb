Heart Disease Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
heart_df = pd.read_csv('heart.csv')
heart_df.head()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(heart_df.corr(), annot=True, cmap='coolwarm')
plt.title("Heart Disease Feature Correlation")
plt.show()
```

```python
# Target distribution
sns.countplot(data=heart_df, x='condition')
plt.title("Heart Disease Cases (0: No, 1: Yes)")
plt.show()
```

Diabetes Dataset Analysis

```python
diabetes_df = pd.read_csv('diabetes.csv')
diabetes_df.head()
```

```python
# Countplot of outcome
sns.countplot(x='Outcome', data=diabetes_df)
plt.title("Diabetes Prediction Distribution")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(diabetes_df.corr(), annot=True, cmap='YlGnBu')
plt.title("Diabetes Feature Correlation")
plt.show()
```

Symptom2Disease Analysis

```python
symptom_df = pd.read_csv('Symptom2Disease.csv')
print(symptom_df.head())
# Top 10 most common diseases
top_diseases = symptom_df['disease'].value_counts().head(10)
top_diseases.plot(kind='barh', color='skyblue')
plt.title("Top 10 Most Common Predicted Diseases")
plt.xlabel("Frequency")
plt.show()
```
